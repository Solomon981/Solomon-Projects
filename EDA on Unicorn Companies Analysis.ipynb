```python
# The top 3 Cities with Highest Concentration of Unicorns

Unicorn_data['Company'] = pd.to_numeric(Unicorn_data['Company'], errors='coerce')
Top3_Cities = Unicorn_data.groupby('City').agg({'Company': 'sum', 'Industry': 'count'}).reset_index()
Top3_Cities.columns = ['City', 'Total Companies','Company']
Top3_Cities = Top3_Cities.sort_values('Company', ascending=False).head(3)
Top3_Cities
```

```python
# Visualize the top 3 cities by Company Concentration
explode = (0, 0, 0)
plt.pie(Top3_Cities['Company'],labels = Top3_Cities['City'], explode=explode, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.title('Top 3 Cities with Highest Company Concentration Across the World')
plt.show()
```

## Observation
- The highest concentration of Unicorn Companies were in the City of San Francisco and New York both in the USA, with 47.8% and 32.3% respectively, making these cities industrial hub for Unicorn Companies.
- Beijing in China has a Unicorn Concentration of 19.9%, making it the third city with high concentration of Unicorns. 

##Industries with Highest Valuing Companies accross the world

```python
## Industries with Highest Valuing Companies accross the world
Top_Industry_Valuation = Unicorn_data.groupby('Industry')['Absolute Valuation'].sum().reset_index()
Top_Industry_Valuation = Top_Industry_Valuation.sort_values('Absolute Valuation', ascending=False).head(3)
Top_Industry_Valuation.columns = ['Industry', 'Total Valuation']
print(Top_Industry_Valuation)

```

```python
# Using donut chart to visualize the Industries with the Highest Valuation

# Define data and explode values
data = Top_Industry_Valuation['Total Valuation']
labels = Top_Industry_Valuation['Industry']
explode = (0, 0, 0)

# Create a donut chart
fig, ax = plt.subplots(figsize=(5, 5))
ax.pie(data, labels=labels, autopct='%1.1f%%', pctdistance=0.85, startangle=90, explode=explode, wedgeprops={'width': 0.5})

# Draw a circle
circle = plt.Circle((0, 0), 0.2, fc='white')
fig.gca().add_artist(circle)

# Add legends
plt.legend(labels, loc='upper left')

# Set title and show chart
plt.title('Top 3 Industries with the highest valuation across the world')
plt.show()
```

## Observation 
- The Fintech Industry has Unicorn Companies with highest valuation  across the world with 46% of Unicorns having high valuation.
- Internet software and service industries have the second largest number of Unicorns with high valuation of 31.4%, followed by E-commerce and direct-to-consumer industries with 22.5%

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Group the data by country and sum the total valuation
Valuation_Country = Unicorn_data.groupby('Country')['Absolute Valuation'].sum()

# Get the top 10 countries by valuation
Top_10_Valuation_Countries = Valuation_Country.sort_values(ascending=False).head(10)

# Group the data by country and count the number of unique investors
Investor_Count_by_Country = Unicorn_data.groupby('Country')['Select Investors'].nunique().sort_values(ascending=False).head(10)

# Abbreviate country names
Top_10_Valuation_Countries.index = ['USA', 'CHN', 'GBR', 'IND', 'DEU', 'FRA', 'CAN', 'KOR', 'SWE', 'JPN']
Investor_Count_by_Country.index = ['USA', 'GBR', 'CHN', 'IND', 'DEU', 'FRA', 'CAN', 'SWE', 'ISR', 'AUS']

# Create a figure with two subplots, one for valuation and one for select investors
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# Plot the data for valuation as a horizontal bar chart
ax1.barh(Top_10_Valuation_Countries.index[::-1], Top_10_Valuation_Countries.values[::-1])
ax1.set_xlabel('Absolute Valuation')
ax1.set_ylabel('Country')
ax1.set_title('Top 10 Countries with Highest Unicorn Valuation')

# Format x-axis tick labels
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: ''.format(x/1e6)))

# Remove the data label for Absolute Valuation
ax1.bar_label(ax1.containers[0], labels=None, label_type='edge')

# Pad the spacing between the number and edge of the figure
ax1.margins(x=0.1)

# Plot the data for select investors as a horizontal bar chart
ax2.barh(Investor_Count_by_Country.index[::-1], Investor_Count_by_Country.values[::-1])
ax2.set_xlabel('Number of Select Investors')
ax2.set_ylabel('Country')
ax2.set_title('Top 10 Countries With the Highest Select Investors')

# Annotate the bars with their values
labels = [str(label) for label in Investor_Count_by_Country.values][::-1]
ax2.bar_label(ax2.containers[0], labels=labels, label_type='edge')

# Pad the spacing between the number and edge of the figure
ax2.margins(x=0.1)

# Show the visual
plt.show()


```

## Observation
- The top 10 Countries with highest number of select investors also produced the Unicorns with high valuations.
- The funding support by Select investors play a significant role in the growth and expansion of Unicorn Companies across the world.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Group the data by country and sum the total valuation
Valuation_Country = Unicorn_data.groupby('Country')['Absolute Valuation'].sum()

# Get the top 10 countries by valuation
Top_10_Valuation_Countries = Valuation_Country.sort_values(ascending=False).head(10)

# Group the data by country and count the number of unique investors
Investor_Count_by_Country = Unicorn_data.groupby('Country')['Select Investors'].nunique().sort_values(ascending=False).head(10)

# Abbreviate country names
Top_10_Valuation_Countries.index = ['USA', 'CHN', 'GBR', 'IND', 'DEU', 'FRA', 'CAN', 'KOR', 'SWE', 'JPN']
Investor_Count_by_Country.index = ['USA', 'GBR', 'CHN', 'IND', 'DEU', 'FRA', 'CAN', 'SWE', 'ISR', 'AUS']

# Create a figure with two subplots, one for valuation and one for select investors
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# Plot the data for valuation as a line graph
ax1.plot(Top_10_Valuation_Countries.index, Top_10_Valuation_Countries.values)
ax1.set_xlabel('Country')
ax1.set_ylabel('Absolute Valuation')
ax1.set_title('Top 10 Countries with Highest Unicorn Valuation in Billions')

# Format y-axis tick labels
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:.2f}'.format(y/1e6)))

# Annotate the data points with their values
for i, val in enumerate(Top_10_Valuation_Countries.values):
    ax1.annotate('{:.2f}'.format(val/1e6), (Top_10_Valuation_Countries.index[i], val))

# Plot the data for select investors as a line graph
ax2.plot(Investor_Count_by_Country.index, Investor_Count_by_Country.values)
ax2.set_xlabel('Country')
ax2.set_ylabel('Number of Select Investors')
ax2.set_title('Top 10 Countries With the Highest Select Investors')

# Annotate the data points with their values
for i, val in enumerate(Investor_Count_by_Country.values):
    ax2.annotate(str(val), (Investor_Count_by_Country.index[i], val))

# Show the visual
plt.show()
```

## Observations
- There is a direct relationship between high funding by Select Investors in Unicorn Companies and the total number of Unicorns Companies  a Country have.

- The higher the number of Select investors in a Country, the higher the number of Unicorn Companies from that Country.
